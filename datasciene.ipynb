```python
dataframe["stroke"].fillna(dataframe["stroke"].mean()  ,inplace=True  )
dataframe["horsepower"].fillna(dataframe["horsepower"].mean() ,inplace=True)
dataframe["peak-rpm"].fillna(dataframe["peak-rpm"].mean ,inplace=True)
dataframe["horsepower-binned"].fillna("Unknown", inplace=True)
```

```python
dataframe.isnull().sum()
```

```python
dataframe['city-L/100km'] = 235/dataframe["city-mpg"]
dataframe['highway-L/100km'] = 235/dataframe["highway-mpg"]
dataframe.drop(columns=["city-mpg"], inplace=True)
dataframe.drop(columns=["highway-mpg"], inplace=True)
dataframe.head()
```

```python
dataframe["length"] = dataframe["length"]/dataframe["length"].max()
dataframe["width"] = dataframe["width"]/dataframe["width"].max()
dataframe["height"] = dataframe["height"]/dataframe["height"].max() 

dataframe[["length","width","height"]].head()
```
